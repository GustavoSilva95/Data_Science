No final de 2019 começou a surgir vários casos de pneumonia na cidade de Wuhan, China. Eram muitos casos aparecendo de repente, os médicos da região não conseguiam identificar a causa desta pneumonia, indicando que havia algo específico causando essa doença.

No dia 31 de dezembro do mesmo ano, a Organização Mundial da Saúde (OMS) emitiu um alerta sobre essa infecção misteriosa. Nos primeiros dias de 2020 a China anunciou o fechamento do mercado de Wuhan, onde supostamente havia começado as infecções. Isso porque aparentemente alguns animais, que tem a carne vendida nesse mercado, eram portadores do vírus que estava causando a pneumonia. Descobriu-se então que esse vírus fazia parte do grupo coronavírus, mais especificamente o SARS-CoV-2.

A doença causada pelo coronavírus SARS-CoV-2 ficou conhecida como COVID-19. Uma infecção respiratória aguda, potencialmente grave e de elevada transmissibilidade. Devido ao fato desta última característica, rapidamente a doença se espalhou pelo mundo, causando uma epidemia global. Não há, na história contemporânea, registro de patologia com evolução tão explosiva, podendo ser transmitida pelo ar, tosse, espirro e até mesmo um aperto de mão. O que levou a OMS, no dia 30 de janeiro de 2020, a decretar estado de emergência global. Em pouco tempo os países estavam decretando a restrição à circulação de pessoas.

Graças ao empenho de pesquisadores ao redor do mundo, compartilhando instantaneamente achados sobre a doença, em novembro de 2020 começam a ser divulgados os primeiros resultados positivos da vacina em testes com humanos.

Assim, o presente projeto tem o objetivo de trazer, utilizando análise exploratória realizada em python, os números da pandemia de Covid-19 através de dados públicos da doença.

### Preparação do Dataset

```python
# Colocando a coluna date como índice do Dataset
df_covid.set_index('date', inplace=True)
```

```python
# Analizando quantos valores únicos possui na coluna continent
df_covid['continent'].unique()
```

Notando a ausência de dados na coluna `continent` faz-se necessário fazer uma verificação destes dados no intuito de encontrar um modo conceitual de agrupar esses dados faltantes. Uma hipótese seria utilizar a coluna `location`, que refere-se ao país, para tentar preencher o `continent`, visto que sabendo o nome do país pode-se inferir o seu continente.

```python
# Verificando os valores únicos de location para os dados faltantes em continent
continent_null = df_covid.loc[df_covid['continent'].isna()]
continent_null['location'].unique()
```

Com a análise acima, foi visto que não será possível fazer o preenchimento dos dados faltantes em `continent`, pois os valores que se esperava encontrar em `location` para esses dados faltantes era o nome dos países. Ao invés disso, foi encontrado nomes de continentes e informações sobre renda. Isso acontece porque a OWID adicionou entradas para totais parciais (continentes) e o total mundial. Com isso um preenchimento conceitual não será possível.

Portanto, para esta análise, o dataset sera dividido em dois. Um não irá conter as entradas da coluna `continent` que estão faltantes, e será chamado de `df_covid_countries`. O outro irá conter somente os dados relacionados a `World`, presentes na coluna `location` e que se referem ao acumulado mundial, chamando `df_covid_world`.

```python
# Criando um Dataset sem as entradas nulas de continent
df_covid_countries = df_covid.dropna(subset=['continent'], axis=0)

# Criando um Dataset com os dados relacionados somente a World
df_covid_world = df_covid.loc[df_covid['location'] == 'World']
```

## Análise Exploratória dos Dados

###COVID-19 no Mundo

```python
# Quantidade de casos de COVID-19 no mundo
total_cases_world = df_covid_world.loc[final]['total_cases']

# colocando ponto de milhar
total_cases_world = '{:,}'.format(int(total_cases_world)).replace(',','.')
print(f'No mundo já foram registrados {total_cases_world} casos')
```

No mundo, já são mais de 281 milhões casos confirmados de Covid-19 atualmente. O número real pode ser ainda maior em consequência de casos assintomáticos ou mais brandos da doença, fazendo com que esses infectados não procurem um hospital. Apesar da baixa taxa de mortalidade, mas devido a elevada transmissibilidade do vírus, o número absoluto de óbitos demonstra a gravidade da pandemia. O gráfico mostra a evolução no número de mortes:

```python
# Gráfico com a evolução do número de mortes por COVID-19 no mundo
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14,8))

#descobrindo quando foi registrado a primeira morte no mundo por COVID-19
start_deaths = df_covid_world.loc[df_covid_world['new_deaths']!=0].head(1).index.tolist()

#descobrindo a data em que se chegou a aproximadamente 1 milhão de mortes
q1 =  df_covid_world.loc[df_covid_world['total_deaths'] >= 1000000.0]['total_deaths'].head(1).tolist()
date_q1 = df_covid_world.loc[df_covid_world['total_deaths'] == q1[0]].index.tolist()

#descobrindo a data em que se chegou a aproximadamente 2 milhão de mortes
q2 =  df_covid_world.loc[df_covid_world['total_deaths'] >= 2000000.0]['total_deaths'].head(1).tolist()
date_q2 = df_covid_world.loc[df_covid_world['total_deaths'] == q2[0]].index.tolist()

#usando o for para parâmetros que são compartilhados por ambos os gráficos
for ax in (ax1, ax2):
  max = '{:,}'.format(int(df_covid_world['total_deaths'].max())).replace(',','.')#total de mortes
  ax.fill_between(df_covid_world.index, df_covid_world['total_deaths'], color='darkred', alpha=0.2)#preenche a área abaixo da curva
  ax.text(x=final, y=df_covid_world['total_deaths'].max(),s=max, fontsize=13, weight='bold')#coloca o valor acumulado
  ax.text(x=date_q1[0], y=q1[0], s='{:,}'.format(int(q1[0])).replace(',','.'), fontsize=13)#coloca o valor próximo de 1 milhão
  ax.text(x=date_q2[0], y=q2[0], s='{:,}'.format(int(q2[0])).replace(',','.'), fontsize=13)#coloca o valor próximo de 2 milhão

  #eliminando as bordas e configurando os ticks
  ax.set_yticklabels([])
  ax.set_yticks([])
  ax.tick_params(axis='x', rotation=45)
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)

  #criando as linhas verticais tracejadas
  ax.axvline(x=date_q2[0], linestyle='dotted', linewidth=2.2, color='steelblue')
  ax.axvline(x=date_q1[0], linestyle='dotted', linewidth=2.2, color='steelblue')

#grafico de escala linear
ax1.plot(df_covid_world['total_deaths'], lw='2', color='darkred')
ax1.text(x=date_q1[0], y=4000000,
         s=f'{(date_q2[0] - date_q1[0]).days} dias\n<------->',
         fontsize=13, color='darkolivegreen')
ax1.text(x=start_deaths[0], y=2500000,
         s=f'        {(date_q1[0] - start_deaths[0]).days} dias\n<------------------->',
         fontsize=13, color='darkolivegreen')
ax1.set_title('Escala linear',fontsize=16, loc= 'left', color='gray')

#gráfico de escala logarítmica
ax2.plot(df_covid_world['total_deaths'], lw='2', color='darkred')
ax2.set_yscale('log')
ax2.text(x=date_q1[0], y=30000,
         s=f'{(date_q2[0] - date_q1[0]).days} dias\n<------->',
         fontsize=13, color='darkolivegreen')
ax2.text(x=start_deaths[0], y=20,
         s=f'        {(date_q1[0] - start_deaths[0]).days} dias\n<------------------->',
         fontsize=13, color='darkolivegreen')
ax2.set_title('Escala logarítmica',fontsize=16, loc= 'left', color='gray')

plt.suptitle(f'Evolução do número de mortes de COVID-19 no mundo até {final.day}/{final.month}/{final.year}',
             color='darkgray', fontsize=24, weight='bold')

plt.show()
```

O mundo já registra mais de 5,4 milhões de mortes pela doença. A partir da primeira morte levou 239 dias para atingir 1 milhão de óbitos, e menos da metade desse tempo, exatamente 115 dias, para dobrar o número de mortes.


#### Países com o maior número de mortes

Os Estados Unidos lideram o ranking, com 818.371 óbitos registradas. Em seguida vêm o Brasil com 618.797 e a Índia com 480.290 mortes. O ranking e o gráfico abaixo mostram os países com o maior número de mortes registradas na pandemia até o momento:

```python
# definir o número de posições no ranking
num_pos = int(input('Digite o n° de posições no ranking: '))
ranking = []
for r in range(1, num_pos + 1):
  r = str(r)
  ranking.append(r + '°')
```

```python
# Ranking dos países com maior número de mortes por COVID-19 (valor absoluto)
ranking_total_deaths = df_covid_countries.loc[final].sort_values(by='total_deaths', ascending=False)[:num_pos]
ranking_total_deaths = ranking_total_deaths[['location', 'total_deaths']]
ranking_total_deaths['ranking'] = ranking
ranking_total_deaths.set_index('ranking', inplace=True)
ranking_total_deaths.head(num_pos)
```

```python
# Gráfico do ranking dos países com maior número de mortes por COVID-19 (valor absoluto)
fig, ax = plt.subplots(figsize=(14, 8))

mortes_absosuto = ax.barh('location', 'total_deaths', data=ranking_total_deaths, color='darkred')
ax.invert_yaxis()
for p in mortes_absosuto.patches:
  plt.annotate('{:,}'.format(int(p.get_width())).replace(',','.'),
              (p.get_width(), p.get_y() + p.get_height() / 2),
              ha='right' ,va='center', fontsize=13, color='white', weight='bold')
ax.set_frame_on(False)
ax.set_xticklabels([])
ax.set_xticks([])
ax.tick_params(axis="y", labelsize=15)
ax.set_title(f'Países com maior número de mortes por COVID-19\naté {final.day}/{final.month}/{final.year}',
             color='darkgray', fontsize=24, weight='bold')
plt.show()
```

#### Análise sobre as variantes da COVID-19

Como todos os vírus, o SARS-CoV-2 tambeém muda com o tempo. Algumas mudanças podem ser imperceptíveis, porém, outras podem afetar propriedades do vírus, como transmissibilidade, imunidade a vacina, entre outras. Por isso, a OMS monitora a evolução do SARS-CoV-2 desde janeiro de 2020 e disponibiliza uma tabela com as Variantes de Interesse (VOIs) e Variantes de Preocupação (VOCs). A tabela mostra as variantes classificadas como VOC e a data em que se tornaram variantes de preocupação.

| Variante | Data VOC |
|:------|:--------|
| alfa | Dezembro de 2020 
| beta | Dezembro de 2020 
| gama | Janeiro de 2021
| delta | Abril de 2021
| omicron | Junho de 2021

A seguir será feita a análise das variantes.

```python
# Grafico de novos casos e mortes a cada variante já identificada
variants = {
    'alpha_beta': '2020-03-01',
    'gamma': '2021-01-11',
    'delta': '2021-05-11',
    'omicron': '2021-11-26',
    }
for (k, v) in variants.items():
  variants[k] = datetime.strptime(v, '%Y-%m-%d').date()

fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols= 1 , figsize=(14, 8))

ax1.plot(df_covid_world['new_cases_smoothed_per_million'], lw=2, color='darkblue')

ax2.plot(df_covid_world['new_deaths_smoothed'], lw=2, color='darkred')

for ax in (ax1,ax2):
  ax.fill_between(df_covid_world.index, df_covid_world['new_cases_smoothed_per_million'] if ax == ax1 else df_covid_world['new_deaths_smoothed'] ,
                 where=(df_covid_world.index < variants['gamma']), color='grey',alpha=0.2)#preenche a área abaixo da curva
  ax.fill_between(df_covid_world.index, df_covid_world['new_cases_smoothed_per_million'] if ax == ax1 else df_covid_world['new_deaths_smoothed'],
                 where=(df_covid_world.index < variants['delta']), color='grey',alpha=0.2)#preenche a área abaixo da curva
  ax.fill_between(df_covid_world.index, df_covid_world['new_cases_smoothed_per_million'] if ax == ax1 else df_covid_world['new_deaths_smoothed'],
                 where=(df_covid_world.index < variants['omicron']), color='gray', alpha=0.2)#preenche a área abaixo da curva
  ax.fill_between(df_covid_world.index, df_covid_world['new_cases_smoothed_per_million'] if ax == ax1 else df_covid_world['new_deaths_smoothed'],
                 where=(df_covid_world.index >= variants['omicron']),color = 'red', alpha=0.2)#preenche a área abaixo da curva

  ax.text(x=variants['omicron'],    y=30 if ax==ax1 else 30*100, s='Omicron', fontsize=13, color='red')
  ax.text(x=variants['delta'],      y=30 if ax==ax1 else 30*100, s='                   Delta', fontsize=13, color='dimgray')
  ax.text(x=variants['gamma'],      y=30 if ax==ax1 else 30*100, s='          Gama', fontsize=13, color='dimgray')
  ax.text(x=variants['alpha_beta'], y=30 if ax==ax1 else 30*100, s='            Alpha + Beta', fontsize=13, color='dimgray')

  ax.tick_params(axis='x', rotation=25)
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.spines['left'].set_visible(False)
  ax.grid(axis='y', linestyle='dotted')
ax1.set_title(f'Novos Casos & Mortes por COVID-19 até {final.day}/{final.month}/{final.year}',
              fontsize=24, color='darkgray', weight='bold')
ax1.text(x=inicial, y=82, s='Média móvel de Novos Casos\n(por milhão)', fontsize=15, color='darkblue')
ax2.text(x=inicial, y=12500, s='Média móvel de Novas Mortes', fontsize=15, color='darkred')

plt.show()
```

O gráfico faz uma comparação do número de casos e mortes no mundo, de acordo com as variantes já identificadas do SARS-CoV-2.

Analisando o gráfico, até o momento, a variante Ômicron se mostra menos letal. Visto que, a média móvel de mortes não acompanha a tendência de alta da média móvel de casos. Esta característica não ocorreu no comportamento das outras variantes, onde a medida que aumentava o número de casos a curva de mortes seguia a mesma tendência
